# Machine Learning, AI e GenAI Tassonomia e Framework MLOps

## 1. Concetti e terminologia fondamentali

### Concetti fondamentali di Machine Learning

- **Machine Learning (ML)**: campo di studio che fornisce ai computer la capacità di apprendere senza essere programmati in modo esplicito
- **Algoritmo**: procedura passo passo per risolvere un problema o portare a termine un'attività
- **Modello**: rappresentazione matematica appresa dai dati per fare previsioni o decisioni
- **Caratteristica**: proprietà o caratteristica misurabile individuale utilizzata come input per un algoritmo di machine learning
- **Etichetta/Target**: valore di output che un modello è addestrato a prevedere
- **Addestramento**: processo di insegnamento di un modello utilizzando dati etichettati
- **Inferenza**: processo di utilizzo di un modello addestrato per fare previsioni
- **Apprendimento supervisionato**: addestramento di un modello su dati etichettati
- **Apprendimento non supervisionato**: ricerca di modelli in dati non etichettati
- **Apprendimento semi-supervisionato**: addestramento con una combinazione di dati etichettati e non etichettati
- **Apprendimento per rinforzo**: apprendimento tramite interazione con un ambiente utilizzando ricompense e penalità
- **Apprendimento per trasferimento**: applicazione delle conoscenze da un modello addestrato a un'attività diversa ma correlata
- **Fine-tuning**: ulteriore addestramento di un modello pre-addestrato su un set di dati specifico per un'attività specifica
- **Sovraadattamento**: quando un modello apprende troppo bene i dati di addestramento, inclusi rumore e valori anomali
- **Sottoadattamento**: quando un modello è troppo semplice per catturare i modelli sottostanti nei dati
- **Bias**: errore dovuto a ipotesi eccessivamente semplicistiche nel modello
- **Varianza**: errore dovuto a troppa complessità nel modello
- **Iperparametro**: set di parametri prima dell'inizio dell'addestramento (non appreso dai dati)
- **Parametro**: valore appreso durante l'addestramento del modello dai dati

### Concetti di intelligenza artificiale

- **Intelligenza artificiale (IA)**: ampio campo che comprende l'apprendimento automatico e i sistemi progettati per imitare l'intelligenza umana
- **AI ristretta/debole**: sistemi di AI progettati per compiti specifici (tutti gli attuali sistemi di AI)
- **AI generale/forte**: AI ipotetica con intelligenza generale simile a quella umana in tutti i domini
- **Intelligenza artificiale generale (AGI)**: AI con la capacità di comprendere, apprendere e applicare la conoscenza in diversi domini
- **Superintelligenza**: AI ipotetica che supera l'intelligenza umana
- **Sistema esperto**: sistema di AI basato su regole che emula l'esperienza umana in un dominio specifico
- **Cognitive Computing**: sistemi di AI che tentano di simulare i processi di pensiero umano
- **Computer Vision**: campo di AI focalizzato sul consentire ai computer di vedere, identificare ed elaborare immagini
- **Natural Language Processing (NLP)**: campo di AI focalizzato sul consentire ai computer di comprendere e generare il linguaggio umano

### Concetti di AI generativa

- **AI generativa (GenAI)**: sistemi di AI in grado di generare nuovi contenuti (testo, immagini, audio, video, ecc.)
- **Modello generativo**: modello in grado di generare nuovi dati istanze simili ai dati di training
- **Large Language Model (LLM)**: rete neurale addestrata su vasti dati di testo in grado di generare testo simile a quello umano
- **Foundation Model**: modello di grandi dimensioni addestrato su dati estesi che può essere adattato a molte attività downstream
- **Diffusion Model**: tipo di modello generativo che aggiunge gradualmente e quindi rimuove rumore per generare dati
- **Prompt**: testo di input che indirizza o istruisce un sistema AI a generare output specifici
- **Prompt Engineering**: pratica di progettazione di prompt efficaci per ottenere gli output desiderati dai sistemi AI
- **In-context Learning**: capacità dei modelli di apprendere dagli esempi forniti nel prompt
- **Retrieval-Augmented Generation (RAG)**: tecnica che migliora la generazione recuperando informazioni rilevanti
- **Allucinazione**: quando l'AI genera informazioni di fatto errate o inesistenti
- **Multimodal Model**: modello in grado di elaborare e generare più tipi di dati (testo, immagine, audio)
- **Text-to-Image**: AI modelli che generano immagini in base a descrizioni di testo
- **Testo in audio/voce**: modelli di intelligenza artificiale che convertono il testo in audio parlato
- **Testo in video**: modelli di intelligenza artificiale che generano contenuti video in base a descrizioni di testo

### Concetti di reti neurali

- **Rete neurale**: sistema di elaborazione ispirato alle reti neurali biologiche
- **Apprendimento profondo**: sottoinsieme di ML che utilizza reti neurali con più livelli
- **Neurone/Nodo**: unità di calcolo di base in una rete neurale
- **Peso**: parametro associato alle connessioni tra neuroni
- **Funzione di attivazione**: funzione che determina l'output di un neurone
- **Backpropagation**: algoritmo per l'aggiornamento dei pesi in base al gradiente di errore
- **Gradient Descent**: algoritmo di ottimizzazione per ridurre al minimo l'errore
- **Epoca**: un passaggio completo attraverso l'intero set di dati di addestramento
- **Batch**: sottoinsieme di dati di addestramento elaborati insieme
- **Livello**: gruppo di neuroni che elaborano l'input e producono l'output
- **Livello nascosto**: livello tra gli strati di input e output
- **Convolutional Neural Network (CNN)**: Rete neurale specializzata per l'elaborazione delle immagini
