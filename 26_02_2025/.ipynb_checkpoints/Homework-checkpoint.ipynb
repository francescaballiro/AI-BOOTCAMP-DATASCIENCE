# EXERCISE 1
Write a Python function that determines if the distribution of a pandas Series is symmetric, right-skewed or left-skewed.

```python
np.random.seed(42)

n = 1000

data = {
    'Age': np.random.randint(18, 70, n),
    'Salary': np.abs(np.random.normal(50000, 15000, n)),
    'Hours_Worked_Per_Week': np.random.randint(20, 60, n),
    'Years_of_Experience': np.random.lognormal(mean=1, sigma=0.7, size=n)
}

df = pd.DataFrame(data)
```

### PRIMO CASO
Per vedere se una funzione è simmetrica posso confrontare media e mediana, se i due valori sono uguali la distribuzione sarà simmetrica

```python
def analizza_simmetria(series):
    # Calcolo la media e la mediana
    media = series.mean()
    mediana = series.median()
    
    # Determino la simmetria in base alla relazione tra media e mediana
    if media > mediana:
        return "Asimmetrica verso destra"
    elif media < mediana:
        return "Asimmetrica verso sinistra"
    else:
        return "Simmetrica"
```

```python
risultato = analizza_simmetria(df['Years_of_Experience'])
print(risultato)
```

### SECONDO CASO
un altro modo in cui posso fare è analizzare la distanza delle code con la mediana

```python
def analizza_simmetria_1(series, lower_percentile=10, upper_percentile=90):
    # Calcolo i percentili specificati (ad esempio, 10° e 90° percentile)
    p1 = series.quantile(lower_percentile / 100)  # Percentile inferiore (es. 10° percentile)
    p2 = series.quantile(upper_percentile / 100)  # Percentile superiore (es. 90° percentile)
    
    # Calcola la mediana
    mediana = series.median()

    # Estraggo i valori al di sotto del 10° percentile e sopra il 90° percentile
    coda_sinistra = series[series < p1]
    coda_destra = series[series > p2]

    # Calcolo la larghezza delle code e la distanza dalla mediana
    larghezza_sinistra = mediana - coda_sinistra.mean()
    larghezza_destra = coda_destra.mean() - mediana

    # Confronto le larghezze delle code
    if larghezza_sinistra == larghezza_destra:
        return "Simmetrica"
    elif larghezza_sinistra < larghezza_destra:
        return "Asimmetrica verso destra"
    else:
        return "Asimmetrica verso sinistra"
```

```python
risultato = analizza_simmetria_1(df['Years_of_Experience'])
print(risultato)
```

### TERZO CASO
uso i percentili per calcolare la distanza con la mediana

```python
def analizza_simmetria_2(series):
    q1 = series.quantile(0.10)
    q3 = series.quantile(0.90)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    mediana = series.median()
    larghezza_sinistra = mediana - lower_bound
    larghezza_destra = upper_bound - mediana

    if larghezza_sinistra == larghezza_destra:
        return "Simmetrica"
    elif larghezza_sinistra < larghezza_destra:
        return "Asimmetrica verso destra"
    else:
        return "Asimmetrica verso sinistra"
```

```python
risultato = analizza_simmetria_2(df['Years_of_Experience'])
print(risultato)
```

# EXERCISE 2
 A disease affects 2% of a population. The population is divided into three age groups:
	•	Young (50%): False positive rate 3%, sensitivity 95%.
	•	Middle-aged (30%): False positive rate 4%, sensitivity 90%.
	•	Elderly (20%): False positive rate 5%, sensitivity 85%.

    N.B.
    False positive rate: probability of a person testing positive when they do not have the disease.
    Sensitivity/True positive rate: probability of a person testing positive when they have the disease.

    If a random person from the population tests positive, what is the probability they actually have the disease?

```python
# Probabilità di avere la malattia e di non averla
p_malat = 0.02  # 2% della popolazione ha la malattia
p_no_malat = 1 - p_malat  # Probabilità di non avere la malattia (98%)

# Probabilità di essere in una fascia d'età
p_giovani = 0.50  # 50% della popolazione è giovane
p_adulti = 0.30  # 30% della popolazione è di mezza età
p_anziani = 0.20  # 20% della popolazione è anziana

# Sensibilità e tasso di falsi positivi per ciascuna fascia d'età
sensibilita_giovani = 0.95  # Sensibilità nei giovani (95%)
falsi_positivi_giovani = 0.03  # Tasso di falsi positivi nei giovani (3%)

sensibilita_adulti = 0.90  # Sensibilità nella fascia di mezza età (90%)
falsi_positivi_adulti = 0.04  # Tasso di falsi positivi nella fascia di mezza età (4%)

sensibilita_anziani = 0.85  # Sensibilità negli anziani (85%)
falsi_positivi_anziani = 0.05  # Tasso di falsi positivi negli anziani (5%)
```

### PRIMO CASO
Calcolo la probabilità complessiva per l'intera popolazione, considerando tutti i gruppi di età insieme.

```python

#la probabilità che una persona risulti positiva al test (p_pos) è calcolata come la somma dei veri positivi  e dei falsi positivi, per ciascun gruppo di età.
p_pos = (
    # Per i giovani
    (sensibilita_giovani * p_malat * p_giovani) + 
    (falsi_positivi_giovani * p_no_malat * p_giovani) +
    
    # Per la mezza età
    (sensibilita_adulti * p_malat * p_adulti) + 
    (falsi_positivi_adulti * p_no_malat * p_adulti) +
    
    # Per gli anziani
    (sensibilita_anziani * p_malat * p_anziani) + 
    (falsi_positivi_anziani * p_no_malat * p_anziani)
)

#calcolo la probabilità globale che una persona abbia la malattia, dato che il test è positivo, considerando l'intera popolazione, quindi combinando le probabilità di tutti i gruppi di età.

p_malatt_given_pos = (
    (sensibilita_giovani * p_malat * p_giovani) + 
    (sensibilita_adulti * p_malat * p_adulti) + 
    (sensibilita_anziani * p_malat * p_anziani)
) / p_pos


print(f"La probabilità che una persona abbia la malattia, dato che il test è positivo, è: {p_malatt_given_pos:.3f}")
```

### SECONDO CASO
calcolo la probabilità separatamente per ogni gruppo di età, ma uso la probabilità totale di un test positivo per l'intera popolazione.

```python
# Calcoliamo la probabilità complessiva (per tutta la popolazione) che tiene conto di tutte le fasce di età messe insieme.
p_pos = (
    # Per i giovani
    (sensibilita_giovani * p_malat * p_giovani) + 
    (falsi_positivi_giovani * p_no_malat * p_giovani) +
    
    # Per la mezza età
    (sensibilita_adulti * p_malat * p_adulti) + 
    (falsi_positivi_adulti * p_no_malat * p_adulti) +
    
    # Per gli anziani
    (sensibilita_anziani * p_malat * p_anziani) + 
    (falsi_positivi_anziani * p_no_malat * p_anziani)
)  

# le probabilità di ciascun gruppo vengono calcolate singolarmente, ma usando la stessa probabilità totale di un test positivo per tutta la popolazione. 
#sto considerando l’intera popolazione, ma con i calcoli separati per ciascun gruppo.
p_malatt_given_pos_giovani = (sensibilita_giovani * p_malat * p_giovani) / p_pos
p_malatt_given_pos_adulti = (sensibilita_adulti * p_malat * p_adulti) / p_pos
p_malatt_given_pos_anziani = (sensibilita_anziani * p_malat * p_anziani) / p_pos



#Ora posso creare un grafico
categorie = ['Giovani', 'Adulti', 'Anziani']
probabilita = [p_malatt_given_pos_giovani, p_malatt_given_pos_adulti, p_malatt_given_pos_anziani]

# Convertiamo le probabilità in percentuali
percentuali = [p * 100 for p in probabilita]

# Creiamo il grafico a barre
plt.figure(figsize=(10, 8))
plt.bar(categorie, percentuali, color=['blue', 'green', 'red'])

# Aggiungiamo etichette e titolo
plt.xlabel('Fascia di età')
plt.ylabel('Probabilità (%) di malattia (dato test positivo)')
plt.title('Probabilità di avere la malattia per fascia di età (dato test positivo)', pad=70)

# Aggiungiamo le percentuali sopra le barre
for i, valore in enumerate(percentuali):
    plt.text(i, valore + 2, f'{valore:.1f}%', ha='center', va='bottom', fontsize=12)

# Mostriamo il grafico
plt.tight_layout()
plt.show()
```

### TERZO CASO
Calcolo la probabilità separatamente per ogni gruppo, utilizzando la probabilità di un test positivo specifica per ogni gruppo di età.

```python
# la probabilità che una persona di ciascun gruppo risulti positiva al test. 
# Calcolo questo valore solo tenendo conto della sensibilità e dei falsi positivi per ciascun gruppo, senza mescolare i gruppi.
p_pos_giovani =(sensibilita_giovani * p_malat * p_giovani) + (falsi_positivi_giovani * p_no_malat * p_giovani)
p_pos_adulti =(sensibilita_adulti * p_malat * p_adulti) + (falsi_positivi_adulti * p_no_malat * p_adulti)
p_pos_anziani =(sensibilita_anziani * p_malat * p_anziani) + (falsi_positivi_anziani * p_no_malat * p_anziani)

# La probabilità viene quindi calcolata separatamente per i giovani, adulti e anziani, utilizzando il rispettivo p_pos per ciascun gruppo.
# Per ogni gruppo, la probabilità di avere la malattia dato che il test è positivo viene calcolata solo considerando quel gruppo di età e la sua specifica probabilità di test positivo.
p_malatt_given_pos_giovani = (sensibilita_giovani * p_malat * p_giovani) / p_pos_giovani
p_malatt_given_pos_adulti = (sensibilita_adulti * p_malat * p_adulti) / p_pos_adulti
p_malatt_given_pos_anziani = (sensibilita_anziani * p_malat * p_anziani) / p_pos_anziani

#Ora posso creare un grafico
categorie = ['Giovani', 'Adulti', 'Anziani']
probabilita = [p_malatt_given_pos_giovani, p_malatt_given_pos_adulti, p_malatt_given_pos_anziani]

# Convertiamo le probabilità in percentuali
percentuali = [p * 100 for p in probabilita]

# Creiamo il grafico a barre
plt.figure(figsize=(10, 8))
plt.bar(categorie, percentuali, color=['blue', 'green', 'red'])

# Aggiungiamo etichette e titolo
plt.xlabel('Fascia di età')
plt.ylabel('Probabilità (%) di malattia (dato test positivo)')
plt.title('Probabilità di avere la malattia per fascia di età (dato test positivo)', pad=70)

# Aggiungiamo le percentuali sopra le barre
for i, valore in enumerate(percentuali):
    plt.text(i, valore + 2, f'{valore:.1f}%', ha='center', va='bottom', fontsize=12)

# Mostriamo il grafico
plt.tight_layout()
plt.show()
```
