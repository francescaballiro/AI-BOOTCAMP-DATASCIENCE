## EXERCISE 1
Write a query to find first name and last name of employees whose first name:
- has 3 letters
- ends in 'a'

```python
%%sql

SELECT first_name, last_name
FROM employees
WHERE LENGTH(first_name) = 3 AND first_name LIKE '%a';
```

```python
%%sql

SELECT 
    d.name AS department_name,
    COUNT(e.id) AS num_employees,
    COUNT(CASE WHEN e.bonus IS NOT NULL THEN 1 END) AS num_employees_with_bonus,
    SUM(CASE WHEN e.bonus IS NOT NULL THEN e.bonus ELSE 0 END) AS total_bonus
FROM departments d
LEFT JOIN employees e ON d.id = e.department_id
GROUP BY d.name;
```

## EXERCISE 3
Bonus exercise

Write a query to find employees who received the same bonus as at least one other employee.

```python
%%sql 

SELECT e1.first_name, e1.last_name, e1.bonus
FROM employees e1
WHERE e1.bonus IN (
    SELECT e2.bonus
    FROM employees e2
    WHERE e2.bonus IS NOT NULL
    GROUP BY e2.bonus
    HAVING COUNT(e2.id) > 1
);
```
