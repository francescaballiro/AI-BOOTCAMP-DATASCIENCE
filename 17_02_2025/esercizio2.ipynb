# Exercise: Create a column to split the data in Low, Medium, High fare prices and calculate the average Age per each section. Fill the Age set to nan and calculate the average age again, how did it change? How can you fill the age so that the average does not change?

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic_1.csv')
df.isnull()
```

Calcolo i quantili al 33%, 66% e 100% per i prezzi delle tariffe

```python
quantiles = df['Fare'].quantile([0, 0.33, 0.66, 1]).values
quantiles
```

Suddividere i dati in "Basso", "Medio" e "Alto" in base ai quantili

```python
def fare_category(fare):
    if fare <= quantiles[1]:
        return 'Basso'
    elif fare <= quantiles[2]:
        return 'Medio'
    else:
        return 'Alto'
```

Applico la funzione alla colonna Fare per creare la nuova colonna

```python
df['Fare_Category'] = df['Fare'].apply(fare_category)
```

```python
df
```

Calcolare l'età media per ciascuna categoria

```python
age_mean = df.groupby('Fare_Category')['Age'].mean()
print(age_mean)
```

Creo una copia del DataFrame e rimuove le righe con valori nulli

```python
df_no_nulls = df.dropna().copy()
df_no_nulls.head()
```

calcolo l'età per ogni categoria cancellando i valori nulli

```python
age_mean = df_no_nulls.groupby('Fare_Category')['Age'].mean()
print(age_mean)
```

per fare in modo che la media non cambi posso riempire i valori NaN con la media della categoria tariffaria a cui appartengono

```python
df['Age'] = df.groupby('Fare_Category')['Age'].transform(lambda x: x.fillna(x.mean()))
```

verifico se l'età media è cambiata

```python
age_mean_fill = df.groupby('Fare_Category')['Age'].mean()
print(age_mean_fill)
```
